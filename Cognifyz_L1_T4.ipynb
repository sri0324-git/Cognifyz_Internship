Online Delivery

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"/content/Dataset .csv")
df.head()
```

```python
df.info()
```

Determine the percentage of restaurants
that offer online delivery.

```python
total_onlinedelivery = len(df[df['Has Online delivery'] == 'Yes'])
total_restaurants = len(df)
percentage_onlinedelivery = (total_onlinedelivery / total_restaurants) * 100
print('the percentage of restaurants that offer online delivery: ,',percentage_onlinedelivery)
```

Compare the average ratings of restaurants
with and without online delivery.

```python
# with online delivery
restaurant_with_onlinedelivery = df[df['Has Online delivery'] == 'Yes']
average_rating_with_onlinedelivery = restaurant_with_onlinedelivery['Aggregate rating'].mean()
print('the average ratings of restaurants with online delivery: ,',average_rating_with_onlinedelivery)
```

```python
# without online delivery
restaurant_without_onlinedelivery = df[df['Has Online delivery'] == 'No']
average_rating_without_onlinedelivery = restaurant_without_onlinedelivery['Aggregate rating'].mean()
print('the average ratings of restaurants without online delivery: ,',average_rating_without_onlinedelivery)
```

```python
labels =['With Online Delivery', 'Without Online Delivery']
average_ratings=[average_rating_with_onlinedelivery, average_rating_without_onlinedelivery]
colors =['green', 'red']
plt.bar(labels, average_ratings, color=colors)
plt.title("The Comparison of Average Ratings of Restaurants with and Without Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Ratings")
plt.show()
```
