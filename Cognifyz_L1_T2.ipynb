CITY ANALYSIS


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"/content/Dataset .csv")
df.head()
```

```python
df.shape
df.info()
```

Identify the city with the highest number
of restaurants in the dataset.

```python
print(df['City'].value_counts()[[0]].index[0])
```

Calculate the average rating for
restaurants in each city.

```python
average_restaurant_rating = df.groupby('City').agg(Average_rating=('Aggregate rating','mean')).sort_values(by='Average_rating',ascending=False)
print(average_restaurant_rating)
```

```python
plt.subplots(figsize=(20,5))
average_restaurant_rating=df.groupby('City').agg(Average_rating=('Aggregate rating','mean')).sort_values(by='Average_rating',ascending=False)
bars=plt.bar(average_restaurant_rating.index,average_restaurant_rating['Average_rating'])
for bar in bars:
  plt.text(bar.get_width(),bar.get_y()+bar.get_height()/2,round(bar.get_width(),2),ha='left',va='center')
plt.title('Average Rating of Restaurants in Each City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.show()
```

Determine the city with the highest
average rating.

```python
average_restaurant_rating=df.groupby('City').agg(Average_rating=('Aggregate rating','mean')).sort_values(by='Average_rating',ascending=False)
print('the city with the highest average rating.',average_restaurant_rating.index[0])
```
